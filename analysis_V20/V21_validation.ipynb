```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# === CONFIG ===
results_dir = Path("../projects/simple_run/results/cow_runV21")
probe_ids = ["p1", "p2", "p10", "p11", "p12"]  # pick any probes you want
filename = "p_mid.txt"  # mid-point pressure

# === LOAD AND PLOT ===
plt.figure(figsize=(12, 6))
for pid in probe_ids:
    file_path = results_dir / pid / filename
    try:
        data = np.loadtxt(file_path)
        time = data[:, 0]  # first column: time
        pressure = data[:, 1]  # second column: pressure
        plt.plot(time, pressure, label=pid)
    except Exception as e:
        print(f"Failed to load {file_path}: {e}")

plt.title("Midpoint Pressure Signals")
plt.xlabel("Time [s]")
plt.ylabel("Pressure [mmHg] (gauge)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
